5.  วันธรรมดาหรือวันเสาร์-อาทิตย์มากกว่ากัน  (วันจันทร์ = 1, วันอังคาร = 2, ...)

````python
monday = merged_table['วันในสัปดาห์'] == 1
tuesday = merged_table['วันในสัปดาห์'] == 2
wednesday =  merged_table['วันในสัปดาห์'] == 3
thursday =  merged_table['วันในสัปดาห์']== 4
friday = merged_table['วันในสัปดาห์']== 5
normal = merged_table[monday | tuesday | wednesday | thursday | friday]
normal
````

````python
normal.groupby('วันในสัปดาห์').count()
````

````python
Ans_nm = normal.groupby('วันในสัปดาห์').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) # เลือก sort_values by column ที่ไม่มี missing มาใช้
Ans_nm
````

````python
sum_nm = Ans_nm[['รหัสสนามบินปลายทาง']].sum()
sum_nm
````

````python
saturday = merged_table['วันในสัปดาห์'] == 6
sunday = merged_table['วันในสัปดาห์'] == 7
holiday = merged_table[saturday | sunday]
holiday
````

````python
holiday.groupby('วันในสัปดาห์').count()
````

````python
Ans_hd = holiday.groupby('วันในสัปดาห์').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) # เลือก sort_values by column ที่ไม่มี missing มาใช้
Ans_hd
````

````python
sum_hd = Ans_hd[['รหัสสนามบินปลายทาง']].sum()
sum_hd
````

ความแตกต่างของวันธรรมดากับวันในสัปดาห์

````python
sum_nm - sum_hd
````

จากโจทย์ วันธรรมดา หรือวันเสาร์-อาทิตย์มีเครื่องบินขาเข้ามากกว่ากัน

วันธรรมดา มีจำนวนเครื่องบินขาเข้า เท่ากับ 2,011,453 ลำ

วันเสาร์-อาทิตย์ มีจำนวนเครื่องบินขาเข้า เท่ากับ 690,765 ลำ
